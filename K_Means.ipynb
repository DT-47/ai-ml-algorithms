# Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
```

# Data Preprocessing

```python
data = pd.read_csv("dataset2.csv")
data.head()
```

```python
data = data.sample(frac=1.0) # Shuffle dataset
```

```python
plt.figure(figsize=(15, 10))
sns.scatterplot(data, x='x', y='y', hue='color')
```

# Clustering

```python
model = KMeans(n_clusters=3)
model.fit(data[["x", "y"]])

predictions = model.predict(data[["x", "y"]])
centers = model.cluster_centers_
print(f"{predictions=}\n{centers=}")
```

```python
from scipy.spatial import Voronoi, voronoi_plot_2d

plt.figure(figsize=(15, 10))
sns.scatterplot(data=data, x='x', y='y', hue=predictions)
plt.scatter(centers[:, 0], centers[:, 1], color='red', s=100)

vor = Voronoi(centers)
fig = voronoi_plot_2d(vor,plt.gca())
plt.show()
```
